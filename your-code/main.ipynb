## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# Import scipy.stats for uniform distribution
from scipy.stats import uniform

# Method 1: Using scipy.stats.uniform
# Generate uniform random numbers between 'a' and 'b'
# uniform.rvs(loc=a, scale=b-a, size=n)
# where loc is the lower bound, scale is the range (upper - lower), size is number of samples

# Example: Generate 5 random numbers between 2 and 8
a = 2  # lower bound
b = 8  # upper bound
n = 5  # number of samples

random_numbers = uniform.rvs(loc=a, scale=b-a, size=n)
print(f"Random numbers between {a} and {b}: {random_numbers}")

# Method 2: Using numpy (alternative approach)
# You can also use numpy.random.uniform directly
random_numbers_numpy = np.random.uniform(low=a, high=b, size=n)
print(f"Random numbers using numpy: {random_numbers_numpy}")
```

```python
# your code here
# Example: Generate 5 random numbers between 2 and 8
a  # lower bound
b # upper bound
n  # number of samples
def gen_random_n(a, b, n):
    random_numbers = uniform.rvs(loc=a, scale=b-a, size=n)
    print(f"Random numbers between {a} and {b}: {random_numbers}")
    return random_numbers

set_one = gen_random_n(10, 15, 100)
set_two = gen_random_n(10, 60, 1000)
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go

import pandas as pd

# Combine the data into a DataFrame for Plotly
df_plot = pd.DataFrame({
    "Value": np.concatenate([set_one, set_two]),
    "Set": (["10-15, n=100"] * len(set_one)) + (["10-60, n=1000"] * len(set_two))
})

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Distribution 1: 10-15, n=100", "Distribution 2: 10-60, n=1000"),
    x_title="Value",
    y_title="Count"
)

# Add first histogram
fig.add_trace(
    go.Histogram(x=set_one, nbinsx=10, name="10-15, n=100", marker_color="blue"),
    row=1, col=1
)

# Add second histogram
fig.add_trace(
    go.Histogram(x=set_two, nbinsx=10, name="10-60, n=1000", marker_color="red"),
    row=1, col=2
)

# Update layout
fig.update_layout(
    title_text="Uniform Distributions Comparison - Side by Side",
    showlegend=True,
    height=400,
    width=900
)

fig.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
import random
# Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
# your code here
def generate_normal_n(mu, sigma, count):
    normal_numbers = []
    for i in range(count):
        number = random.normalvariate(mu, sigma)
        normal_numbers.append(number)

    return normal_numbers

mu = 10
sigma = 1
count = 1000

normal_set_1 = generate_normal_n(mu, sigma, count)
print(f"Generated {(count)} samples with a avg of {mu} and a standard deviation of {sigma}")
print(f"Print 10 first numbers: {normal_set_1[:10]}")

# 1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
sigma = 50
normal_set_2 = generate_normal_n(mu, sigma, count)
print(f"\nGenerated {(count)} samples with a avg of {mu} and a standard deviation of {sigma}")
print(f"Print 10 first numbers: {normal_set_2[:10]}")
    
```

```python
# your code here
# 2. Plot the distributions of the data generated.
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=("Normal Distribution 1", "Normal Distribution 2"))

fig.add_trace(
    go.Histogram(x=normal_set_1, name="μ=10, σ=1", marker_color="green"),
    row=1, col=1
)

fig.add_trace(
    go.Histogram(x=normal_set_2, name="μ=10, σ=50", marker_color="orange"),
    row=1, col=2
)

fig.update_layout(height=600, width=800, title_text="Normal Distribution Comparison", showlegend=True)

fig.update_xaxes(title_text="Value")
fig.update_yaxes(title_text="Frequency")

fig.show()
```

```python
from scipy.stats import skew, kurtosis

def test_normality(data, variable_name):
    """
    Comprehensive normality test for a dataset
    """
    import scipy.stats as stats
    import matplotlib.pyplot as plt
    
    print(f"=== Normality Tests for {variable_name} ===")
    
    # 1. Descriptive statistics
    print(f"Skewness: {skew(data):.4f} (Normal ≈ 0)")
    print(f"Kurtosis: {kurtosis(data):.4f} (Normal ≈ 0)")
    
    # 2. Shapiro-Wilk test
    if len(data) <= 5000:  # Shapiro works best for smaller samples
        stat, p = shapiro(data)
        print(f"Shapiro-Wilk p-value: {p:.6f}")
    
    # 3. Visual tests
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Histogram
    ax1.hist(data, bins=30, density=True, alpha=0.7, edgecolor='black')
    ax1.set_title(f'Histogram of {variable_name}')
    
    # Q-Q plot
    stats.probplot(data, dist="norm", plot=ax2)
    ax2.set_title(f'Q-Q Plot of {variable_name}')
    
    plt.tight_layout()
    plt.show()

# Use the function
test_normality(vehicles_df['Fuel Barrels/Year'], 'Fuel Barrels/Year')
```

```python
# your code here
def random_numbers_exponential(mu, size):
    return np.random.exponential(scale=mu, size=size)

exp_10 = random_numbers_exponential(10, 10)
exp_100 = random_numbers_exponential(10, 100)
print(exp_10)
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)
def exp_dis_formula(lam, e, x):
    lam
    e = math.e
    x
    F = 1 - (e ** (-lam * x))
    return F


lam = 0.1
e = math.e
x = 15
result_exp_dis_formula = exp_dis_formula(lam, e, x)
print(f"The probability that a customer will spend less than fifteen minutes is: {result_exp_dis_formula:.1%}")
```
